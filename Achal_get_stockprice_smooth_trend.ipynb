```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
df_google_01=df_google.close.to_list()
type(df_google_01)
```

```python
avg=[]
window_size=50
i=0
for i in range(len(df_google)-window_size):
    df_google_break=df_google_01[i:window_size+i]
    temp=sum(df_google_break)/window_size
    avg.append(temp)
last_n_list=df_google_01[-window_size:]
temp2=sum(df_google_break)/window_size
a_list_01=[temp2]*window_size
moving_average=avg+a_list_01
```

```python
df_google["moving_average"]=moving_average
df_google.plot(x="date",y="moving_average",figsize=(16,8))
plt.show()
```
